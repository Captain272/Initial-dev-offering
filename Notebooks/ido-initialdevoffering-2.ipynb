# Pre process

lowercase ---
Remove stop words ---
implement porter ---
remove " none "

```python
import pandas as pd
```

```python
df=pd.read_csv("../input/gituserdata/final.csv")

df= pd.concat([df,df2])
 
MULTIPLE_WHITESPACE = re.compile('\s+')
email_pattern = re.compile('\S*@\S*\s?')
keys = dict.fromkeys(string.punctuation)
for k in keys:keys[k] = " "
table = str.maketrans(keys)
 
def preprocessor(x):
    x = x.lower()
    x = email_pattern.sub('email', x)
    x = re.sub('www\S+|http://\S+|https://\S+', 'url', x)
    x = x.translate(table)
    x = MULTIPLE_WHITESPACE.sub(" ", x).strip()
    x=x.replace('none', '')
    return x

df.content=df.content.astype(str)
df['content'] = df['content'].apply(preprocessor)
```
